# III.I) Multi-Modality Multi-Task Model

#### ‚ñ∂ Using Multi-Modailty Teqniques because we have two kind of data (text and Images) so we want to compine the result of the two models in one model so we use Multi-modality Technique 

‚≠ê In **Trial #1** we will use **Multi-modality** that use Two kinds of data(Text and Images) with **Multi-Task** to predict **both Price and Type** of the houses in our dataset.

‚≠ê In this Trial we will use simple **average of embedding** to train tesxt data and will use simple **conv2D**, **Maxpool2D** and **flatten** Layers to train the images, after that we will concat the result of two models and then use **two Dense layers** with **Softmax** activation function **one for each Task**.

## III.I.I) Trial #1 (Using Multi-Modality & Multi-Task)

```python
# Importing Libraries
from __future__ import absolute_import, division, print_function, unicode_literals
import collections
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPool2D
from tensorflow.keras.optimizers import Adam


# here we have two inputs. one for image and the other for text.
in_text = keras.Input(batch_shape=(None, max_len))
in_image = keras.Input(batch_shape=(None, 64, 64, 2))

# text part
# simple average of embedding.
embedded = keras.layers.Embedding(tokenizer.num_words, 100)(in_text)
averaged = tf.reduce_mean(embedded, axis=1)


# image part 
# simple conv2d. 
cov = Conv2D(32, (16, 16))(in_image)
pl = MaxPool2D((16, 16))(cov)
flattened = Flatten()(pl)


# fusion - combinig both
fused = tf.concat([averaged, flattened], axis=-1)

# multi-task learning (each is a multi-class classification)
# one dense layer for each task
p_type = Dense(len_type, activation='softmax', name='type')(fused)
p_price = Dense(len_price, activation='softmax', name='price')(fused)


# define model input/output using keys.
model = keras.Model(
    inputs={
        'summary': in_text,
        'image': in_image
    },
    outputs={
        'type': p_type,
        'price': p_price,
    },
)


# compile model with optimizer, loss values for each task, loss 
model.compile(
    optimizer=Adam(),
    loss={
        'type': 'sparse_categorical_crossentropy',
        'price': 'sparse_categorical_crossentropy',
    },
    loss_weights={
        'type': 0.5,
        'price': 0.5,       
    },
    metrics={
        'type': ['SparseCategoricalAccuracy'],
        'price': ['SparseCategoricalAccuracy'],
    },
)


model.summary()
```

```python
# Fitting the model
history = model.fit(
    x={
        'summary': x_train_text_id,
        'image': x_train_image
    },
    y={
        'type': y_train_type,
        'price': y_train_price,
    },
    epochs=20,
    batch_size=16,
    validation_split=0.2,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_price_loss', patience=5, )
    ],
    verbose=1
)
```

```python
# we can do prediction on Testing set
y_predict = model.predict(
    {
        'summary': x_test_text_id,
        'image': x_test_image
    }
)


# probabilities
price_predicted = y_predict['price']
print(price_predicted)

# categories
price_category_predicted = np.argmax(price_predicted, axis=1)
print(price_category_predicted)
```

## III.I.II) Trial #2 (Using Droupout Layer)

‚≠ê In **Trial #2** we will use **Multi-modality** that use Two kinds of data(Text and Images) with **Multi-Task** to predict **both Price and Type** of the houses in our dataset.

‚≠ê In this Trial we will use simple **average of embedding** to train text data and will use simple **conv2D**, **Maxpool2D** and **flatten** Layers to train the images, after that we will concat the result of two models and then we add **Two Droupout layers** and then use **two Dense layers** with **Softmax** activation function **one for each Task**. 

‚≠ê Then we will compile the model with **Adam** optimizer and using the **SparseCategoricalAccuracy** loss function and **SparseCategoricalAccuracy** metric to measure the performance of the model.

```python
# Importing Libraries
from __future__ import absolute_import, division, print_function, unicode_literals
import collections
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPool2D
from tensorflow.keras.optimizers import Adam


# here we have two inputs. one for image and the other for text.
in_text = keras.Input(batch_shape=(None, max_len))
in_image = keras.Input(batch_shape=(None, 64, 64, 2))

# text part
# simple average of embedding. 
embedded = keras.layers.Embedding(tokenizer.num_words, 100)(in_text)
averaged = tf.reduce_mean(embedded, axis=1)


# image part 
# simple conv2d.
cov = Conv2D(32, (16, 16))(in_image)
pl = MaxPool2D((16, 16))(cov)
flattened = Flatten()(pl)


# fusion - combinig both
fused = tf.concat([averaged, flattened], axis=-1)

# Adding Dropout Layer
fused1 = Dense(128, activation='relu')(fused)
dropout_1 = Dropout(rate = 0.2)(fused1)
fused1 = Dense(256, activation='relu')(dropout_1)


fused2 = Dense(256, activation='relu')(fused)
dropout_2 = Dropout(rate = 0.3)(fused2)
fused2 =  Dense(128, activation='relu')(dropout_2)

# multi-task learning (each is a multi-class classification)
# one dense layer for each task
p_type = Dense(len_type, activation='softmax', name='type')(fused1)
p_price = Dense(len_price, activation='softmax', name='price')(fused2)


# define model input/output using keys.
model = keras.Model(
    inputs={
        'summary': in_text,
        'image': in_image
    },
    outputs={
        'type': p_type,
        'price': p_price,
    },
)


# compile model with optimizer, loss values for each task, loss 
model.compile(
    optimizer=Adam(),
    loss={
        'type': 'sparse_categorical_crossentropy',
        'price': 'sparse_categorical_crossentropy',
    },
    loss_weights={
        'type': 0.5,
        'price': 0.5,       
    },
    metrics={
        'type': ['SparseCategoricalAccuracy'],
        'price': ['SparseCategoricalAccuracy'],
    },
)


model.summary()
```

```python
# Fitting the model
history = model.fit(
    x={
        'summary': x_train_text_id,
        'image': x_train_image
    },
    y={
        'type': y_train_type,
        'price': y_train_price,
    },
    epochs=20,
    batch_size=16,
    validation_split=0.2,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_price_sparse_categorical_accuracy', patience=5, )
    ],
    verbose=1
)
```

```python
# we can do prediction on Testing set
y_predict = model.predict(
    {
        'summary': x_test_text_id,
        'image': x_test_image
    }
)


# probabilities
price_predicted = y_predict['price']
print(price_predicted)

# categories
price_category_predicted = np.argmax(price_predicted, axis=1)
print(price_category_predicted)
```

## üåà **Q1:**  Is fully-connected model a good one for sequential data? Why? How about for image data? Is it good? Why?


‚≠ê A fully-connected model may not be a good choice for sequential data because it does not take into account the sequential nature of the data. Fully-connected models are also known as feedforward neural networks and are typically used for non-sequential data such as tabular data. Sequential data, on the other hand, has a temporal dimension, and the order of the data points matters.

‚≠ê For sequential data, recurrent neural networks (RNNs) such as the GRU and LSTM are more suitable. RNNs are designed to handle sequential data by maintaining a "memory" of the previous inputs, which allows them to capture temporal dependencies and patterns. The GRU and LSTM are both types of RNNs that are commonly used for natural language processing (NLP) tasks such as text classification and language translation.

‚≠ê For image data, fully-connected models may not be the best choice either. Image data is high-dimensional and has a spatial structure that is not captured by fully-connected layers. Instead, convolutional neural networks (CNNs) are typically used for image data. CNNs are designed to handle the spatial structure of image data by using convolutional layers that can learn spatial features such as edges and textures.

üî• In summary, fully-connected models are not typically used for sequential or image data. Instead, RNNs such as the GRU and LSTM are used for sequential data, while CNNs are used for image data. These specialized types of models are designed to capture the unique characteristics of each type of data and achieve better performance than fully-connected models.

## üåà **Q2:** What is gradient vanishing and gradient explosion, and how GRU/LSTM tries to mitigate this problem?


üëâüèª Gradient vanishing and gradient explosion are two common problems that can occur during the training of neural networks, including recurrent neural networks (RNNs) such as the GRU and LSTM.

‚òòÔ∏è Gradient vanishing occurs when the gradients become very small as they propagate through the layers of the network during backpropagation, which can cause the weights to update very slowly or not at all. This can lead to slow convergence or even convergence to a suboptimal solution.

‚òòÔ∏è Gradient explosion, on the other hand, occurs when the gradients become very large, which can cause the weights to update too quickly and lead to overfitting or divergence.

üëâüèª The GRU and LSTM were designed to mitigate these problems by using a mechanism called gated recurrent units (GRUs) or long short-term memory (LSTM) cells, respectively. These mechanisms use gates to control the flow of information through the network and allow it to selectively remember or forget information from previous time steps.

‚≠ê In the GRU, there are two gates: a reset gate and an update gate. The reset gate controls how much of the previous hidden state is used to compute the current hidden state, while the update gate controls how much of the new input is used to update the current hidden state.

‚≠ê In the LSTM, there are three gates: an input gate, a forget gate, and an output gate. The input gate controls how much of the new input is used to update the memory cell, while the forget gate controls how much of the previous memory cell is forgotten. The output gate controls how much of the memory cell is used to compute the current hidden state.

üëâüèª XGBoost, LightGBM, and CatBoost are all popular open-source gradient boosting frameworks used for machine learning tasks. While they share some similarities, there are some differences between them:

‚òòÔ∏è Training speed: LightGBM and CatBoost are generally faster than XGBoost due to their optimized algorithms for handling large datasets. ‚òòÔ∏è Regularization techniques: All three frameworks support various regularization techniques such as L1 and L2 regularization, but CatBoost also supports a novel technique called "ordered boosting" that can handle non-random patterns in the data.

‚òòÔ∏è Categorical feature handling: CatBoost is designed to handle categorical features more efficiently than XGBoost and LightGBM. It uses an ordered boosting algorithm that can handle categorical features with a large number of categories and missing values without requiring feature engineering.

üî• In summary, while XGBoost, LightGBM, and CatBoost share some similarities such as their use of gradient boosting algorithms, they differ in terms of training speed, memory efficiency, categorical feature handling, regularization techniques, and programming language compatibility.
